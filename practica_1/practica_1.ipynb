### 3. Python

Crear en python las siguientes celdas:

- Definir una clase "Calculadora" con métodos "suma" y "multiplicación" que tomen dos parámetros y regresen el resultado de la operación
- Definir una clase "CalculadoraMultiple" con métodos "suma" y "multiplicación" que tomen cualquier cantidad de parámetros y regresen el resultado de la operación 
- Instanciar ambas clases en una nueva celda y comprobar su funcionamiento

```python
# Clase Calculadora
class Calculadora:
    def suma(self, a, b):
        return a + b

    def multiplicacion(self, a, b):
        return a * b
```

```python
# Clase CalculadoraMultiple
class CalculadoraMultiple:
    def suma(self, *args):
        return sum(args)

    def multiplicacion(self, *args):
        resultado = 1
        for num in args:
            resultado *= num
        return resultado
```

```python
# Crear instancias de ambas clases
calc = Calculadora()
calc_multi = CalculadoraMultiple()

# Probar métodos de Calculadora
print("Calculadora (2 parámetros):")
print("5 + 3 =", calc.suma(5, 3))
print("5 * 3 =", calc.multiplicacion(5, 3))

# Probar métodos de CalculadoraMultiple
print("\nCalculadora (múltiples parámetros):")
print("1 + 2 + 3 + 4 =", calc_multi.suma(1, 2, 3, 4))
print("2 * 3 * 4 =", calc_multi.multiplicacion(2, 3, 4))
```

## Análisis de resultados

¿Qué utilidad tiene el uso de un repositorio en línea para código?
> Un repositorio en línea (como GitHub, GitLab o Bitbucket) sirve para almacenar, compartir y gestionar el código de un proyecto en la nube. Sus principales utilidades son:
Respaldo: Evita la pérdida del código al estar guardado en línea.
Colaboración: Permite que varias personas trabajen simultáneamente en el mismo proyecto.
Control de versiones: Guarda el historial de cambios y facilita volver a versiones anteriores.
Integración continua: Facilita automatizar pruebas, despliegues y revisiones.

¿Cuál es la convención de ángulos que utilizan los archivos URDF para los ángulos de las juntas respecto al eslabón padre?
> En los archivos URDF (Unified Robot Description Format), la convención utilizada para representar los ángulos de las juntas respecto al eslabón padre es la convención de Euler extrínseca en el orden Roll-Pitch-Yaw (RPY), que representa rotaciones alrededor de los ejes X, Y y Z, respectivamente.
Roll (φ): Rotación alrededor del eje X
Pitch (θ): Rotación alrededor del eje Y
Yaw (ψ): Rotación alrededor del eje Z

¿Qué utilidad tiene describir un robot en un archivo URDF?
> Describir un robot en un archivo URDF es útil porque permite definir de forma estructurada y estándar la geometría, estructura y propiedades físicas del robot. Esto incluye sus eslabones, juntas, masas, colisiones y visualizaciones.
Las principales utilidades son:
Simulación: Permite usar el modelo en simuladores (como Gazebo) para probar comportamientos sin un robot físico.
Visualización: Facilita ver la estructura del robot en herramientas como RViz.
Control: Los programas pueden interpretar el URDF para planificar movimientos y calcular cinemática/dinámica.
Interoperabilidad: El mismo archivo se puede usar en distintos entornos de desarrollo o plataformas robóticas.
